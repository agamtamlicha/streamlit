# 4. Memisahkan Data Training dan Data Testing
Berdasarkan dataset bersih, kita memisahkannya.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import joblib

df = pd.read_csv('dataset_bersih.csv')
X = df.drop('target', axis=1)
y = df['target']
```

Standarisasi Skala (*Scaling*) agar SVM dapat memproses seimbang.

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
joblib.dump(scaler, 'scaler.pkl')

X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

Membagi porsi 80% Latih dan 20% Uji.

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled_df, y, test_size=0.2, random_state=42)

X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
print('Selesai memisahkan Data Train dan Data Test!')
```
