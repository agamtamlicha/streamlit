# 1. Tentukan Library yang digunakan
Melakukan import untuk library manipulasi data.

```python
import pandas as pd
import numpy as np
```

# 2. Load Dataset
Membaca dataset utama untuk memulai proses pembersihan data.

```python
# Load dataset asli dengan format skiprows Kodim
df = pd.read_csv('data_semapta.csv', skiprows=19, sep=';', header=None, on_bad_lines='skip')
df.head()
```

# 3. Standarisasi Data (Cleaning & Fitur)
Membersihkan koma dan merapikan kolom fitur.

```python
# Ekstraksi fitur fisik penting yang akan diproses AI
fitur_cols = [7, 10, 11, 27, 29, 31, 33, 35]
X_raw = df.iloc[:, fitur_cols].copy()
X_raw.columns = ['umur', 'tb', 'bb', 'lari', 'pullup', 'situp', 'pushup', 'shuttle']

# Target Kelas
y = df.iloc[:, 48].copy()

# Konversi nilai Koma (,) ke Titik (.) agar terbaca numerik
for col in X_raw.columns:
    X_raw[col] = X_raw[col].astype(str).str.replace(',', '.')
    X_raw[col] = pd.to_numeric(X_raw[col], errors='coerce')

# Gabungkan kembali dan simpan untuk tahap berikutnya
df_clean = X_raw.copy()
df_clean['target'] = y
df_clean = df_clean.dropna()

df_clean.to_csv('dataset_bersih.csv', index=False)
print(f'Data bersih siap diproses: {len(df_clean)} sampel')
```
