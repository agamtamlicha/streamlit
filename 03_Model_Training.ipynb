# 5. Membuat data latih menggunakan algoritma SVM
Melatih SVM murni dari file terpisah.

```python
import pandas as pd
from sklearn.svm import SVC
import joblib

X_train = pd.read_csv('X_train.csv')
y_train = pd.read_csv('y_train.csv')

model_svm = SVC(kernel='linear', random_state=42)
model_svm.fit(X_train, y_train['target'])

joblib.dump(model_svm, 'model_svm.pkl')
print('Proses Training Selesai. Otak tersimpan di model_svm.pkl')
```
